

### **Exploring the Impact of Alcohol and Cigarette Consumption on Job Performance and Social Interaction**

### Description:
In this dataset, we explore the effects of alcohol consumption and cigarette smoking on various aspects of employee well-being, including job performance and social interaction. The dataset includes a range of variables to help analyze these relationships, providing insights into how lifestyle choices might influence work and social dynamics.

**Columns in the Dataset:**

- **`Employee_ID`**: A unique identifier for each employee, used to distinguish and track individual records.

- **`Alcohol_Consumption`**: Measures the amount of alcohol consumed per week, in units. This variable helps us understand the impact of alcohol consumption on employee outcomes.

- **`Cigarette_Smoking`**: Indicates the number of cigarettes smoked per day. This data is used to analyze the relationship between smoking habits and various employee metrics.

- **`Job_Performance`**: A score out of 100 reflecting the employee’s performance in their job. This measure provides insight into how alcohol and cigarette use might affect job performance.

- **`Social_Interaction`**: A score out of 100 representing the employee’s level of social interaction. This variable helps us explore the effects of lifestyle factors on social engagement.

- **`Department`**: Categorical variable indicating the department (e.g., HR, Finance, IT, Marketing, Sales) where the employee works. This helps in understanding departmental differences in job performance and social interaction.

- **`Job_Level`**: Indicates the employee's level within the organization (Junior, Mid, Senior). This variable is useful for analyzing how job seniority impacts or interacts with lifestyle factors.

- **`Smoking_Category`**: Categorizes employees based on their smoking habits (Non-Smoker, Light Smoker, Heavy Smoker). This variable helps segment the data by smoking intensity for more detailed analysis.

This dataset provides a rich foundation for analyzing how alcohol and cigarette consumption impact job performance and social interaction, offering valuable insights for workplace well-being and policy-making.

### When to Use Box Plots:
# Comparing Distributions:

Use box plots to compare distributions across different groups. For example, comparing test scores across different classes or departments.
# Identifying Outliers:

Box plots effectively highlight outliers, which are data points that fall outside 1.5 times the interquartile range (IQR).
This helps in understanding the variability in your data.
# Understanding Spread and Central Tendency:

They provide a clear view of the median, quartiles, and overall range of the data. 
This helps in understanding the central tendency and variability.
# Visualizing Non-Normal Distributions:

When your data is not normally distributed, box plots can be a better alternative to histograms, especially when you want to visualize multiple distributions side by side.
# Grouped Data:

When you have categorical variables and want to visualize the distribution of a continuous variable across different categories.

```python
f1 = {'family': 'Times New Roman', 'size' : 12, 'color' : 'green'}
sns.boxplot(y ='Social_Interaction', x = 'Department', data = df, hue = 'Smoking_Category' )
plt.ylim(20,110)
plt.ylabel('Social Interaction')
plt.xlabel('Depertments ', fontdict = f1)
plt.legend(edgecolor = 'r', fontsize = 12, fancybox = True, shadow = True)
plt.tight_layout()
plt.show()
```

```python
# AOB  SWARM PLOT  VS STRIP PLOT
sns.stripplot(y ='Social_Interaction', x = 'Department',data= df)
sns.swarmplot(y ='Social_Interaction', x = 'Department', data = df, hue = 'Smoking_Category' )
plt.ylim(20,110)
plt.ylabel('Social Interaction')
plt.xlabel('Depertments ')
plt.legend(edgecolor = 'r', fontsize = 12, fancybox = True, shadow = True)
plt.tight_layout()
plt.show()
```
